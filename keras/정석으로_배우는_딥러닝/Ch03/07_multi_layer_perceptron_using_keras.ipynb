```python
# Package Loaded
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
np.random.seed(0)
```

```python
X = np.array([ [0,0],
                        [0,1],
                        [1,0],
                        [1,1] ])
Y = np.array([[0],[1],[1],[0]])
```

```python
# Single layer perceptron

model = Sequential()
model.add(Dense(input_dim=2, units=1))
model.add(Activation('sigmoid'))
model.compile(loss='binary_crossentropy', optimizer=SGD(lr=0.1))
history=model.fit(X, Y, epochs=200, batch_size=1)
```

```python
# Single layer 는 학습 실패.
prob = model.predict_proba(X, batch_size=1)
print('probability : ')
print(prob)
```

```python
# Multi Layer perceptron

model = Sequential()
model.add(Dense(input_dim=2, units=2))
model.add(Activation('sigmoid'))
model.add(Dense(units=1))
model.add(Activation('sigmoid'))
model.compile(loss='binary_crossentropy', optimizer=SGD(lr=0.1))
history=model.fit(X, Y, epochs=4000)
```

```python
# Multi Layer 는 학습 준수.
classes = model.predict_classes(X, batch_size=4)
print('classified : ')
print(classes==Y)
print('\n-----------------------------\n')
prob = model.predict_proba(X, batch_size=4)
print('probability : ')
print(prob)
```
